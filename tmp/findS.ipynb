```python
import datetime

# 获取当前日期
today = datetime.date.today()
# 定义周数量
weeks_count = 20

# 计算开始日期
start_date = today - datetime.timedelta(days=today.weekday() + 7 * weeks_count)
end_date  = today

# 格式化日期为字符串
start_date_str = start_date.strftime('%Y%m%d')
end_date_str = end_date.strftime('%Y%m%d')

print(start_date_str)
print(end_date_str)
```

```python
stock_zh_a_hist_df = ak.stock_zh_a_hist(symbol="000001", period="weekly", start_date=start_date_str, end_date=end_date_str, adjust="qfq")

stock_cyq_em_df = ak.stock_cyq_em(symbol="000001", adjust="qfq")

# 计算成交量的平均值
average_volume = stock_zh_a_hist_df['成交量'][:-1].mean()

# 提取最后一行的成交量值
last_volume = stock_zh_a_hist_df.iloc[-1]['成交量']

# 计算周量比
volume_ratio = round(last_volume / average_volume,2)

# 周期最高价
high_max = stock_zh_a_hist_df['最高'][:-1].max()

# 周期最低价
low_min = stock_zh_a_hist_df['最低'][:-1].min()

# 周期收盘价
close = stock_zh_a_hist_df.iloc[-1]['收盘']
# 获取最后一行最近的5个 "收盘" 的平均值
ma5 = stock_zh_a_hist_df['收盘'].tail(5).mean()
ma10 = stock_zh_a_hist_df['收盘'].tail(10).mean()

# 计算价格比例
close_ratio = round((close-low_min)/(high_max-low_min) *100)

stock_comment_em_df.loc[stock_comment_em_df['代码'] == '000001', '周量比'] = volume_ratio
stock_comment_em_df.loc[stock_comment_em_df['代码'] == '000001', '周价比'] = close_ratio
stock_comment_em_df.loc[stock_comment_em_df['代码'] == '000001', 'MA5'] = ma5
stock_comment_em_df.loc[stock_comment_em_df['代码'] == '000001', 'MA10'] = ma10
stock_comment_em_df.loc[stock_comment_em_df['代码'] == '000001', '获利比例'] = stock_cyq_em_df.iloc[-1]['获利比例']
stock_comment_em_df.loc[stock_comment_em_df['代码'] == '000001', '平均成本'] = stock_cyq_em_df.iloc[-1]['平均成本']
stock_comment_em_df.loc[stock_comment_em_df['代码'] == '000001', '计算日'] = end_date_str

# 创建新的DataFrame
#new_df = pd.DataFrame({'股票代码': ['000001'],'周量比': [volume_ratio],'周价比': [close_ratio]})

stock_comment_em_df.head(1)
```

```python
# 将上述内容转为函数

def add_weekly_metrics(symbol):
    # 获取股票历史数据
    stock_zh_a_hist_df = ak.stock_zh_a_hist(symbol=symbol, period="weekly", start_date=start_date_str, end_date=end_date_str, adjust="qfq")
    #stock_cyq_em_df = ak.stock_cyq_em(symbol=symbol, adjust="qfq")
    
    # 计算成交量的平均值
    average_volume = stock_zh_a_hist_df['成交量'][:-1].mean()
    
    # 提取最后一行的成交量值
    last_volume = stock_zh_a_hist_df.iloc[-1]['成交量']
    
    # 计算周量比
    volume_ratio = round(last_volume / average_volume, 2)
    
    # 周期最高价和最低价
    high_max = stock_zh_a_hist_df['最高'][:-1].max()
    low_min = stock_zh_a_hist_df['最低'][:-1].min()
    
    # 周期收盘价
    close = stock_zh_a_hist_df.iloc[-1]['收盘']
    # 获取最后一行最近的5个 "收盘" 的平均值
    ma5 = stock_zh_a_hist_df['收盘'].tail(5).mean()
    ma10 = stock_zh_a_hist_df['收盘'].tail(10).mean()
    
    # 计算价格比例
    close_ratio = round((close - low_min) / (high_max - low_min) * 100)
    
    # 将计算结果添加到 stock_comment_em_df 中
    stock_comment_em_df.loc[stock_comment_em_df['代码'] == symbol, '周量比'] = volume_ratio
    stock_comment_em_df.loc[stock_comment_em_df['代码'] == symbol, '周价比'] = close_ratio
    stock_comment_em_df.loc[stock_comment_em_df['代码'] == symbol, 'MA5'] = ma5
    stock_comment_em_df.loc[stock_comment_em_df['代码'] == symbol, 'MA10'] = ma10
    #stock_comment_em_df.loc[stock_comment_em_df['代码'] == symbol, '获利比例'] = stock_cyq_em_df.iloc[-1]['获利比例']
    #stock_comment_em_df.loc[stock_comment_em_df['代码'] == symbol, '平均成本 '] = stock_cyq_em_df.iloc[-1]['平均成本']
    stock_comment_em_df.loc[stock_comment_em_df['代码'] == symbol, '计算日'] = end_date_str
    
# 测试
add_weekly_metrics("000002")
stock_comment_em_df.head(3)
```
